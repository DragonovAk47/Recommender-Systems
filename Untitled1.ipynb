```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import seaborn as sns
import scipy.io
import h5py
import scipy.optimize as opt
```

```python
data = scipy.io.loadmat('ex8_movies.mat')
```

```python
R = data['R']
```

```python
data = data['Y']
```

```python
features = 10
movies = 1682
users = 944
my_ratings = np.zeros([1682,1])
my_ratings[7,0] = 3
my_ratings[12,0]= 5
my_ratings[54,0] = 4
my_ratings[64,0]= 5
my_ratings[66,0]= 3
my_ratings[69,0] = 5
my_ratings[183,0] = 4
my_ratings[226,0] = 5
my_ratings[355,0]= 5

ratings = data
ratings = np.concatenate((my_ratings,ratings),axis = 1)
r = R
r1 = np.zeros([1682,1])
for i in range(0,movies):
    if my_ratings[i,0]>0:
        r1[i,0] = 1
r = np.concatenate((r1,r),axis = 1)
```

```python
ratings.shape
```

```python
def normalize_ratings(r,ratings,movies,users):
    m  = movies
    Ymean = np.zeros(m)
    Ynorm = np.zeros(ratings.shape)
    for i in range(m):
        idx = np.where(r[i] == 1)
        Ymean[i] = np.mean(ratings[i, idx])
        Ynorm[i, idx] = ratings[i, idx] - Ymean[i]
    return Ynorm,Ymean
        
```

```python
ratings,Ymean = normalize_ratings(r,ratings,movies,users)
X  = np.random.randn(movies,features)
theta = np.random.randn(users,features)
lambd = 1.5
```

```python
def compute_cost(ratings,r,X,theta,lambd,movies,users,features):
    predict = np.dot(X,theta.T)*r - ratings
    cost = (np.sum(predict*predict)+ (lambd)*(np.sum(np.square(X)) + (lambd)*np.sum(np.square(theta))))/2000
    theta_grad = np.dot(X.T,predict*r).T + (lambd)*theta
    X_grad = np.dot(predict*r,theta)  +(lambd)* X
    return cost,theta_grad,X_grad
```

```python
cost,theta_grad,X_grad = (compute_cost(ratings,r,X,theta,0.5,movies,users,features))
```

```python
print(cost)
```

```python
alpha = 0.001
for i in range(0,400):
    cost,theta_grad,X_grad =compute_cost(ratings,r,X,theta,lambd,movies,users,features)
    print(cost)
    theta = theta - (alpha)*(theta_grad)
    X = X - (alpha)*(X_grad)
```

```python
predictions = np.dot(X,theta.T)
Ymean = Ymean.reshape([1682,1])
```
